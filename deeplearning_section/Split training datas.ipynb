```python
import os
import json
import numpy as np
import pandas as pd
from shutil import copyfile
```

```python
data_containing_fld = './Data/Raw/testing/20200416/'
destination_fld = './Data/Training data/raws/'
```

```python
moving_rate = 1.0
```

```python
for category in os.listdir(data_containing_fld):
    category_path = data_containing_fld + category + '/'
    if '.DS' in category_path:
        continue
    destination_path = destination_fld + category + '/'
    if not os.path.exists(destination_path):
        os.mkdir(destination_path)
        
    for file in os.listdir(category_path):
        if '_i.csv' in file:
            rand = np.random.random()
            if rand < moving_rate:
                file_path = category_path + file
                file_dest = destination_path + file
                copyfile(file_path, file_dest)
                print('Copying file {} to {}'.format(file, category_path))
#         else:
#             print('Not intensity file')
```
