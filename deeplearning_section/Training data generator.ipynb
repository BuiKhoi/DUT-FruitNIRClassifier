```python
import numpy as np
import matplotlib.pyplot as plt
import os
import json
from common_processing import *
import pandas as pd
```

```python
X_data = []
y_true = []
y_true_dict = []
count = 0
all_path = './Data/Training data/raws/'
for folder in os.listdir(all_path):
    if '.' not in folder:
        lbl = count
        fld_path = all_path + folder + '/'
        for dirr in os.listdir(fld_path):
            if '_i.csv' in dirr:
                print(fld_path + dirr)
                df = pd.read_csv(fld_path + dirr)
                if len(df.columns) > 2:
                    df = df.loc[:, df.columns.intersection(['Wavelength (nm)', 'Sample Signal (unitless)'])]
                try:
                    dfnp = df.to_numpy()[:, 1]
#                     if dfnp.shape != (228, ):
#                         print(dfnp.shape)
                    pps = preprocess_spectrum(dfnp)
                    X_data.append(pps)
                    y_true.append(lbl)
                except IndexError as e:
                    print(e)
                    print(df)
#                     raise Exception
        count += 1
    y_true_dict.append(folder)
```

```python
X_data = np.asarray(X_data)
y_true = np.asarray(y_true)
print(X_data.shape)
print(y_true.shape)
```

```python
np.savez_compressed('./Data/Training data/training_3layers', xdata = X_data, ytrue = y_true)
```
