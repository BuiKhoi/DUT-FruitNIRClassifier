```python
base = './Data/train/train/Avocado/'
for dirr in os.listdir(base):
    if '.csv' in dirr:
        if '_i' not in dirr:
            with open(base + dirr, 'r') as read_file:
                txtstr = read_file.read()
            txtstr = txtstr[606:]
            with open(base + dirr, 'w') as write_file:
                write_file.write(txtstr)
```

```python
X_data = []
y_true = []
y_true_dict = []
count = 0
all_path = './Data/train/train/'
for folder in os.listdir(all_path):
    if '.' not in folder:
        lbl = count
        fld_path = all_path + folder + '/'
        for dirr in os.listdir(fld_path):
            if '.csv' in dirr:
                print(fld_path + dirr)
                df = pd.read_csv(fld_path + dirr)
                if len(df.columns) > 2:
                    df = df.loc[:, df.columns.intersection(['Wavelength (nm)', 'Sample Signal (unitless)'])]
                try:
                    dfnp = df.to_numpy()[:, 1]
                    if dfnp.shape != (228, ):
                        print(dfnp.shape)
                    X_data.append(np.expand_dims(input_norm(calc_deriative(dfnp)), 1))
                    y_true.append(lbl)
                except IndexError:
                    print(df)
#                     raise Exception
        count += 1
    y_true_dict.append(folder)
```

```python
X_data = np.asarray(X_data)
y_true = np.asarray(y_true)
print(X_data.shape)
print(y_true.shape)
```

```python
np.savez_compressed('./Data/Training data/training_der', xdata = X_data, ytrue = y_true)
```
