```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp = np.load('./Data/Training data/training.npz')
X_data = temp['xdata']
y_true = temp['ytrue']
```

```python
def calc_deriative(sequence):
    deriative = np.empty(len(sequence) - 1)
    for i in range(len(deriative)):
        deriative[i] = sequence[i + 1] - sequence[i]
    return deriative
```

```python
rand = np.random.randint(0, len(X_data))
plt.plot(X_data[rand]/20)
plt.plot(calc_deriative(calc_deriative(X_data[rand])))
```
